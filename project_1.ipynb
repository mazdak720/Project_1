##### 10. Create a dictionary for each column mapping the State to its respective value for that column. 

```python
states_rate = {}
states_verbal = {}
states_math = {}
for row in sat_scores:
    for i, col in enumerate(row):
        if (i == 0):
            header = col
        elif (i == 1):
            states_rate[header] = col
        elif (i == 2):
            states_verbal[header] = col
        else:
            states_math[header] = col

print states_rate            
print states_verbal            
print states_math            
```

##### 11. Create a dictionary with the values for each of the numeric columns

```python
rates_dict = {column_names[1]: []}
verbals_dict = {column_names[2]: []}
maths_dict = {column_names[3]: []}

for row in sat_scores:
    for i, col in enumerate(row):
#        columns[header(i)].append(col)
        if (i == 1):
            rates_dict[column_names[i]].append(col)
        elif (i == 2):
            verbals_dict[column_names[i]].append(col)
        elif (i == 3):
            maths_dict[column_names[i]].append(col)

columns_dict = rates_dict.copy()
columns_dict.update(verbals_dict)
columns_dict.update(maths_dict)

print rates_dict
print verbals_dict
print maths_dict
print columns_dict

```

##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def column_std(column):
    col_len = len(column)
    col_mean = sum(column) / col_len
    squared_diff = [(data - col_mean) ** 2 for data in column]
    col_std = (sum(squared_diff) / col_len) ** 0.5
    return(col_std)

for column in columns_dict.keys():
    print "The standard deviation for " + column + " column is " + str(column_std(columns_dict[column]))
```

##### 19. Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
def scatter_array(data_dict):
    array_size = len(data_dict.keys())
    columns = data_dict.keys()
    f, arr = plt.subplots(array_size, array_size, sharex=True)
    for row in range(array_size):
        column1 = columns[row]
        for col in range(array_size):
            column2 = columns[col]
            arr[row][col].scatter(data_dict[column1], data_dict[column2], c="red", alpha=0.5)
            arr[row][col].set_title('Scatter plot of ' + column1 + " and " + column2 + ":", {'fontsize': 7})
    plt.show()

    
scatter_array(columns_dict)
```
